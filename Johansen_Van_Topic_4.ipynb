## Part A

```python
# Install Libraries and dependencies
import pandas as pd
import numpy as np
from sklearn.datasets import make_blobs
from matplotlib import pyplot
import neurolab as nl
```

```python
# part A Data
# generate 2d classification dataset
X, y = make_blobs(n_samples=1000, center_box=[-2,2], n_features=2, centers=1)
z = []
for i in range(0, len(X)):
    if (X[:,0][i] <= 0):
        z.append(0)
    else:
        z.append(1)
        
# scatter plot, dots colored by class value
df_part_a = pd.DataFrame(dict(x=X[:,0], y=X[:,1], label=z))
colors = {0:'green', 1:'red'}
fig, ax = pyplot.subplots()
grouped = df_part_a.groupby('label')
for key, group in grouped:
    group.plot(ax=ax, kind='scatter', x='x', y='y', label=key, color=colors[key])
pyplot.title('Decision Boundary Part A')
pyplot.show()
```

```python
# Set up data and set up neural net
input_dat = X
target = []
for v in z:
    target.append([v])

# Create net with 2 inputs and 1 neuron
net = nl.net.newp([[0, 1],[0, 1]], 1)
net.layers[-1].transf = nl.trans.HardLim()
```

```python
# Train Neural Net
error = net.train(input_dat[:800], target[:800], epochs=500, goal=0.1)
print("Finish error: ", error[-1])

# Number of network inputs:
print("Number of network inputs: ", net.ci)

# Number of network outputs:
print("Number of network outputs: ", net.co)

# Number of network layers:
print("Number of network layers: ", len(net.layers))

# Weight of first neuron of input layer (net.layers[0])
print("Weight of neuron: ", net.layers[0].np['w'])

# Bias output layer:
print("Bias output layer: ", net.layers[-1].np['b'])

# Run on Test Set - Automation
net.layers[-1].transf = nl.trans.HardLim()

preds = net.sim(input_dat[800:])

# variance calculation
print("Variance: ", np.var(preds))
```
